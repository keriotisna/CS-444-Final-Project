```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getFlops(M, N):
    
    return (2/N + 2/(M*N**2) + 9/(M**2 * N**2))


mValues = list(range(1, 9))
nValues = list(range(1, 9))

mList = []
nList = []
flopsList = []
compressionRatios = []
colors = []

for M in mValues:
    for N in nValues:
        flops = getFlops(M, N)
        mList.append(M)
        nList.append(N)
        flopsList.append(flops)

        threshold = 2/(M*N) + 9/((M*N)**2)

        # Assign color based on threshold
        if flops > threshold:
            colors.append('red')
        else:
            colors.append('blue')

sizes = [f * 100 for f in flopsList]

plt.scatter(mList, nList, s=sizes, c=colors)
plt.xlabel('M Values')
plt.ylabel('N Values')
plt.title('Scatter Plot of Flops for different M and N')
plt.show()


```
